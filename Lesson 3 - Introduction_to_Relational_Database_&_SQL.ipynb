<h1><center>Hours Tracking with Customized Functions</center></h1>

```python
import pandas as pd
from matplotlib import pyplot as plt

def get_weekly_hours(week_of_month):
    weekly_hours = []
    for i in week_of_month:
        task_1 = float(input("Enter " + i + "'s hours for task 1:"))
        task_2 = float(input("Enter " + i + "'s hours for task 2:"))
        task_3 = float(input("Enter " + i + "'s hours for task 3:"))
        task_4 = float(input("Enter " + i + "'s hours for task 4:"))
        weekly_hours.append({'Week':i,'Task 1':task_1,'Task 2':task_2,'Task 3':task_3, 'Task 4':task_4})
    df = pd.DataFrame(weekly_hours) # Assign the list of dictionaries to a dataframe
    return df


def chart_tasks(week_of_month, task_1,task_2,task_3,task_4):
    plt.plot(week_of_month, task_1, color='purple',marker='D')
    plt.plot(week_of_month, task_2, color='orange',marker='s')
    plt.plot(week_of_month, task_3, color='red',marker='p')
    plt.plot(week_of_month, task_4, color='green',marker='*')
    plt.plot(week_of_month, avg, color='brown',linestyle='--')
    plt.legend(['Task 1', 'Task 2','Task 3','Task 4','Avg',]) 
    plt.xlabel("Week of Month")
    plt.ylabel("Hours")
    plt.title('Tasks Comparison')
    plt.show()
    return

def prepare_tasks(task_list):
    week_of_month = []
    task_1 = []
    task_2 = []
    task_3 = []
    task_4 = []
    avg = []
    total = 0
    for i in task_list:
        task_1.append(i[0])
        task_2.append(i[1])
        task_3.append(i[2])
        task_4.append(i[3])
        week_of_month.append(i[4])
        total = float(i[0]) + float(i[1]) + float(i[2]) + float(i[3])
        avg.append(total/4)
    chart_tasks(week_of_month,task_1,task_2,task_3,task_4)
    return
#####################################################################
week_of_month = ['Week 1','Week 2','Week 3','Week 4']   
weekly_hours = get_weekly_hours(week_of_month)
weekly_hours.to_csv('weekly_hours.csv',mode='w',index=False)
task_hours = pd.read_csv("weekly_hours.csv") # import csv file by using the read_csv functon in Pandas
task_list = task_hours.values.tolist()
prepare_tasks(task_list)
df # list the dataframe
```

<center><h1>DDL (Data Definition Language)</h1></center>

```python
import sqlite3  # Import the sqlite3 library into the current Jupyter Notebook

# Before operating on a database, one has to establish a connection to it.
# If the database doesn't exist, sqlite3 will automatically create it.
# Use the built-in cursor function associated with the connection variable (i.e. con) to create a reference to
# the connection (i.e con.cursor()). The function will return the cursor object and we can assign it to a 
# variable. In the following example, the cursor is called "cursor".
# Define SQL commands as two string variables, drop_sql and sql.
# Use the try/except coding structure (kind of like the if/then structure) to process the SQL commands and 
# handle any exception

con = sqlite3.connect('demo1.db') # Connection established.
cursor = con.cursor()            # Built-in cursor function called.
drop_sql = "DROP TABLE IF EXISTS contacts"                                         # SQL command defined
sql = "CREATE TABLE contacts (id INTEGER, name TEXT, age INTEGER, birthday DATE);" # SQL command defined
try:
    cursor.execute(drop_sql)
    cursor.execute(sql)
    con.commit()
    print("Table successfully created.")
except Exception as e:
    print("Exception occurred.",e)
cursor.close()
con.close()
```

```python
import pandas as pd # Import pandas library, which contains a collection of data manipulation modules.
import sqlite3      # Import sqlite3 library, which contains a collection of SQL database modules.  
con = sqlite3.connect('demo.db') # Connection established.
cursor = con.cursor()            # Built-in cursor function called.
sql = "SELECT contacts.name, countries.country_name FROM contacts INNER JOIN countries ON contacts.id = countries.country_id;"  
try:
    cursor.execute(sql)
    con.commit()
    print("Entry successfully retrieved.")
except Exception as e:
    print("Exception occurred.",e)
# Display results
result = cursor.fetchall()
# print(type(result))
# print(result)
# print(pd.DataFrame(result)) 
# print('\n')
con.close()
df = pd.DataFrame(result, columns=[ 
'Student','Country'
  ])
df
```

<center><h1>One-to-Many</h1></center>

```python
import pandas as pd # Import pandas library, which contains a collection of data manipulation modules.
import sqlite3      # Import sqlite3 library, which contains a collection of SQL database modules.  
con = sqlite3.connect('demo.db') # Connection established.
cursor = con.cursor()            # Built-in cursor function called.
sql = "SELECT contacts.id as contact_id, contacts.name, call_log.log_id, call_log.log_date, countries.country_name as country, call_log.notes FROM contacts INNER JOIN call_log ON contacts.id = call_log.contact_id INNER JOIN countries ON contacts.country_id = countries.country_id;"  
try:
    cursor.execute(sql)
    con.commit()
    print("Entry successfully retrieved.")
except Exception as e:
    print("Exception occurred.",e)
# Display results
result = cursor.fetchall()
# print(type(result))
# print(result)
# print(pd.DataFrame(result)) 
# print('\n')
con.close()
df = pd.DataFrame(result, columns=[ 
'Contact Id','Contact Name', 'Log Id','Log Date','Country','Notes'
  ])
df
```

<center><h1>Many-to-Many</h1></center>

```python
import pandas as pd # Import pandas library, which contains a collection of data manipulation modules.
import sqlite3      # Import sqlite3 library, which contains a collection of SQL database modules.  
con = sqlite3.connect('demo.db') # Connection established.
cursor = con.cursor()            # Built-in cursor function called.
sql = "SELECT contacts.name, schools.school_name, contacts_schools.leave_date, contacts_schools.leave_grade FROM contacts JOIN contacts_schools ON contacts_schools.id = contacts.id JOIN schools ON schools.school_id= contacts_schools.school_id;"
try:    
    cursor.execute(sql)
    con.commit()
    print("Entry successfully retrieved.")
except Exception as e:
    print("Exception occurred.",e)
# Display results
result = cursor.fetchall()
# print(type(result))
# print(result)
# print(pd.DataFrame(result)) 
# print('\n')
con.close()
df = pd.DataFrame(result, columns=[ 
'Name','School','Leave Date','Leave Grade/Level'
  ])
df
```
